```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import the csv
x = pd.read_csv('rollercoasters.csv')
```

```python
#Change the csv file into a pandas DataFrame
df = pd.DataFrame(x)
```

```python
#Show the first 5 rows of data to get a sense for what the data look like
df.head()
```

```python
#Create subplots to display the data in the DataFrame.
plt.figure(figsize=(15,15))

#Pictured is a breakdown of the counts of excitement ratings for rollercoasters in the DataFrame. This histogram shows that
#the vast majority of rollercoasters had an excitement rating of "High".
plt.subplot(2, 2, 1)
plt.hist(df['excitement_rating'], color = 'orange')
plt.ylabel('Count of Rollercoasters')
plt.title('Rollercoaster Excitement Ratings')

#This shows the count of excitement ratings using a line graph. I chose to display this graph with the previous one to show the
#that the average rating is slightly higher than half, which complements the count of "High" ratings in the first graph. The x-axis
#represents each of the different rollercoasters in the DataFrame.
plt.subplot(2, 2, 2)
plt.plot(df['excitement'], color = 'purple')
plt.xlabel('Rollercoaster ID #')
plt.ylabel('Excitement Rating')
plt.title('Excitement Rating Line Graph')

#This scatterplot shows the amount of time each rollercoaster takes for a passenger to ride. Unlike the previous graphs, these data
#represent all of the individual parks (x-axis) and each of the rollercoasters within the park.
plt.subplot(2, 2, 3)
plt.scatter(x = df['park_id'], y = df['ride_time'], color = 'g')
plt.xlabel('Park ID #')
plt.ylabel('Ride Time (seconds)')
plt.title('Rollercoaster Ride Time')

#I chose to use a boxplot to represent the number of drops to see if there were any outliers as occured in the other data points.
#It is clear from the boxplot that most rollercoasters have 10 or fewer drops, but there are a handful with more. I was surprised
#to see that one of the rollercoasters has over 40 drops.
plt.subplot(2, 2, 4)
plt.boxplot(df['drops'])
plt.ylabel('Number of Drops')
plt.title('Rollercoaster Drops Boxplot')

plt.show()
```
