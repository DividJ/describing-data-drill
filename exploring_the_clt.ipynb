```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Part 1: reproduce the pop1 and pop2 populations and samples. Specifically, create two binomially distributed populations with n equal to 10 and size equal to 10000. The p-value of pop1 should be 0.2 and the p-value of pop2 should be 0.5. Using a sample size of 100, calculate the means and standard deviations of your samples.
np.random.seed(0)

pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
```

```python
#Plot the samples as histograms 
np.random.seed(0)

plt.figure(figsize = (10, 5))

plt.subplot(1,2,1)
plt.hist(sample1, color = 'red')
plt.xlabel('Mean: {}, StdDev: {}'.format(sample1.mean(), round(sample1.std(), 2)))
plt.title('pop1 with Sample Size of 100 (sample1)')

plt.subplot(1,2,2)
plt.hist(sample2, color = 'yellow')
plt.xlabel('Mean: {}, StdDev: {}'.format(sample2.mean(), round(sample2.std(), 2)))
plt.title('pop2 with Sample Size of 100 (sample2)')

plt.show()
```

```python
#For each of the following tasks, first write what you expect will happen, then code the changes and observe what does happen.

#Question 1: Increase the size of your samples from 100 to 1000, then calculate the means and standard deviations for your new samples and create histograms for each. Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?
    #Initial thoughts: Increasing the sample size from 100 to 1000 should make the data look more like a normal distribution. This would also be true of the means and standard deviations for the populations as well. As the size of the samples increases the closer we should get to the overall populations' true mean and standard deviation. Conversely, lowering the sample size will likely make the data appear more random. The means and standard deviations could look very different from what was calculated above. These calculations would not be as representative of the entire population.

#Increase sample sizes from 100 to 1000
np.random.seed(0)

sample3 = np.random.choice(pop1, 1000, replace=True)
sample4 = np.random.choice(pop2, 1000, replace=True)

#Calculate the means and standard deviations. What changed and what didn't?

#Create histograms for each new sample
plt.figure(figsize = (10, 5))

plt.subplot(1,2,1)
plt.hist(sample3, color = 'maroon')
plt.xlabel('Mean: {}, StdDev: {}'.format(round(sample3.mean(), 2), round(sample3.std(), 2)))
plt.title('pop1 with Sample Size of 1000 (sample3)')

plt.subplot(1,2,2)
plt.hist(sample4, color = 'purple')
plt.xlabel('Mean: {}, StdDev: {}'.format(round(sample4.mean(), 2), round(sample4.std(), 2)))
plt.title('pop2 with Sample Size of 1000 (sample4)')

plt.show()

print('For sample3, compared to the initial sample, the mean is slightly lower and the std dev is slightly higher.')
print('For sample4, compared to the initial sample, the mean decreased and the std dev increased.')
print('\nOverall, the mean and std dev for both new samples remained fairly similar, which is not unusual after the increase in the sample size.')

#Decrease sample sizes to 20
sample5 = np.random.choice(pop1, 20, replace=True)
sample6 = np.random.choice(pop2, 20, replace=True)

#Calculate the means and standard deviations. What changed and what didn't?

#Create histograms for each new sample
plt.figure(figsize = (10, 5))

plt.subplot(1,2,1)
plt.hist(sample5, color = 'teal')
plt.xlabel('Mean: {}, StdDev: {}'.format(sample5.mean(), round(sample5.std(), 2)))
plt.title('pop1 with Sample Size of 20 (sample5)')

plt.subplot(1,2,2)
plt.hist(sample6, color = 'green')
plt.xlabel('Mean: {}, StdDev: {}'.format(sample6.mean(), round(sample6.std(), 2)))
plt.title('pop2 with Sample Size of 20 (sample6)')

plt.show()

print('For sample5, compared to the initial sample, the mean increased and the std dev decreased.')
print('For sample6, compared to the initial sample, both the mean and the std dev decreased.')
print('\nThe change in the mean for both samples is not surprising. I was expecting to see the std dev increase for both samples and was surprised to see that they both decreased.')
```

```python
#Question 2: Change the population value for pop1 to 0.3, then take new samples and compute the t-statistic and p-value.
np.random.seed(0)

pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

diff = sample2.mean() - sample1.mean()
print('The difference between the two sample means is {}.'.format(diff))

size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

diff_se = (sum(sd ** 2 / size)) ** 0.5  

print('\n', 'The calculated t-value is {}.'.format(diff/diff_se))

from scipy.stats import ttest_ind
print('\n', ttest_ind(sample2, sample1, equal_var=False))
```

```python
#Question 2: Then change the population value p for group 1 to 0.4, and do it again.
np.random.seed(0)

pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

diff = sample2.mean() - sample1.mean()
print('The difference between the two sample means is {}.'.format(diff))

size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

diff_se = (sum(sd ** 2 / size)) ** 0.5  

print('\n', 'The calculated t-value is {}.'.format(diff/diff_se))

from scipy.stats import ttest_ind
print('\n', ttest_ind(sample2, sample1, equal_var=False))
```

```python
#Question 2: What changes, and why?

#Each of the calculated values changes pretty significantly. This happens because the probability of the events occuring are now closer than they were before.
```

```python
#Question 3: Change the distribution of your populations from binomial to a distribution of your choice. Do the sample mean values still accurately represent the population values?
np.random.seed(0)

pop1 = np.random.gamma(10, 0.4, 10000)
pop2 = np.random.gamma(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

diff = sample2.mean() - sample1.mean()
print('The difference between the two sample means is {}.'.format(diff))

size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

diff_se = (sum(sd ** 2 / size)) ** 0.5  

print('\n', 'The calculated t-value is {}.'.format(diff/diff_se))

from scipy.stats import ttest_ind
print('\n', ttest_ind(sample2, sample1, equal_var=False))
```
