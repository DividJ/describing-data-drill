```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Question 1: Create 6 random variables with 100 datapoints using different distributions for each

#Binomial Distribution
binomialvar = np.random.binomial(5, 0.6, 100)

#Gamma Distribution
gammavar = np.random.gamma(10, 2, 100)

#Poisson Distribution
poissonvar = np.random.poisson(25, 100)

#Exponential Distribution
exponentvar = np.random.exponential(5, 100)

#Normal Distribution
normalvar = np.random.normal(18, 4, 100)

#Logistic Distribution
logvar = np.random.logistic(18, 4, 100)
```

```python
#Questions 2 and 3: Plot the variables in a histogram and calculate the mean and standard deviation and plot them as vertical lines on the graph. 

plt.figure(figsize=(15, 10))

plt.subplot(2,3,1)
plt.hist(binomialvar, color = 'blue')
plt.axvline(binomialvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomialvar.mean() + binomialvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomialvar.mean()-binomialvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Mean = {}, Std Dev = {}'.format(round(binomialvar.mean(), 2), round(binomialvar.std(), 2)))
plt.title('Binomial Distribution')

plt.subplot(2,3,2)
plt.hist(gammavar, color = 'green')
plt.axvline(gammavar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gammavar.mean() + gammavar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gammavar.mean()-gammavar.std(), color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Mean = {}, Std Dev = {}'.format(round(gammavar.mean(), 2), round(gammavar.std(), 2)))
plt.title('Gamma Distribution')

plt.subplot(2,3,3)
plt.hist(poissonvar, color = 'purple')
plt.axvline(poissonvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poissonvar.mean() + poissonvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poissonvar.mean()-poissonvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Mean = {}, Std Dev = {}'.format(round(poissonvar.mean(), 2), round(poissonvar.std(), 2)))
plt.title('Poisson Distribution')

plt.subplot(2,3,4)
plt.hist(exponentvar, color = 'red')
plt.axvline(exponentvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponentvar.mean() + exponentvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponentvar.mean()-exponentvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Mean = {}, Std Dev = {}'.format(round(exponentvar.mean(), 2), round(exponentvar.std(), 2)))
plt.title('Exponential Distribution')

plt.subplot(2,3,5)
plt.hist(normalvar, color = 'orange')
plt.axvline(normalvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normalvar.mean() + normalvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normalvar.mean()-normalvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Mean = {}, Std Dev = {}'.format(round(normalvar.mean(), 2), round(normalvar.std(), 2)))
plt.title('Normal Distribution')

plt.subplot(2,3,6)
plt.hist(logvar, color = 'yellow')
plt.axvline(logvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logvar.mean() + logvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(logvar.mean()-logvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Mean = {}, Std Dev = {}'.format(round(logvar.mean(), 2), round(logvar.std(), 2)))
plt.title('Logarithmic Distribution')

plt.show()
```

```python
#Question 4a: Evaluate whether the descriptive statistics provided useful information about the variable.
    #The descriptive statistics for the gamma, binomial, poisson, normal, and logarithmic distributions show that the data are normal because the curves that are produced are generally displayed in a normal way.
    #The exponential distribution is less normal.
    
#Question 4b: Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not?
    #The gamma, poisson, normal, and logarithmic distributions have a mean that is shown in the center of the graph and have standard deviations that are more closely aligned with a normal distribution curve.
    #The binomial distribution has a mean in the center of the graph, but has a larger relative standard deviation than the others.
    #The exponential distribution has a mean that is skewed to the left of the graph and also has a larger relative standard deviation than the others. 
```

```python
#Additional Questions:

#Question 1:Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.

rand1 = np.random.normal(5, 0.5, 100)
rand2 = np.random.normal(10, 1, 100)

#Question 2: Add them together to create a third variable.

rand3 = rand1 + rand2

#Question 3: Graph the third variable using a histogram.

plt.hist(rand3, color = 'magenta')
plt.title('Graph of Two Random Variables Added Together')
plt.show()

#Question 4: Compute the mean and standard deviation and plot them as vertical lines on the histogram.

plt.hist(rand3, color = 'red')
plt.axvline(rand3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean()-rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Mean = {}, Std Dev = {}'.format(round(rand3.mean(), 2), round(rand3.std(), 2)))
plt.title('Random Variable 3 with Mean and Std Dev')
plt.show()

#Question 5: Evaluate the descriptive statistics against the data.
    #The output of the two variables added together show a relatively normal distribution, with the mean centered in the graph and around 50% of the data on the data on either side of the mean. Additionally, the data mostly lie within one standard deviation of the mean, which is another indicator that the variable is normal.
```
